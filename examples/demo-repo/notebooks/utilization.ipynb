# Station utilization
Rides per dock, from the demo db. (Registered in `librarian-artifacts.toml`.)

```python
import sqlite3, pathlib
conn = sqlite3.connect(':memory:')
conn.executescript(pathlib.Path('../db/seed.sql').read_text())
list(conn.execute('SELECT s.name, count(*)*1.0/s.dock_count FROM rides r JOIN stations s USING (station_id) GROUP BY s.name ORDER BY 2 DESC LIMIT 5'))
```
